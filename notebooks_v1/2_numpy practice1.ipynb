# Creating Arrays from Python Lists
## - np.array()
## - .shape

```python
# Turn Python list in to Numpy array
np.array([1,3,5,7,9])
```

```python
# Nested list result 5 x 5 Matrix
np.array([range(i,i + 5) for i in [1,2,3,4,5]])
```

```python
# Nested list result 10 x 10 Matrix
np.array([range(i,i+10) for i in range(1,11)])
```

```python
# Count the shape (rows , columns)
np.array([range(i,i+10) for i in range(1,11)]).shape
```

# Creating Arrays from Scratch


```python
# Create a 10 x 10 matrix integer array filled with zeros
np.zeros((10,10), dtype=int)
```

```python
# Create a 10 x 10 matrix of integer filled with ones
np.ones((10,10),dtype=int)
```

```python
# Create a 2 X 3 Matrix of interger filled with 2.5
np.full((2,3),2.5)
```

```python
# Create a 3 x 3 identity matrix
np.eye(3)
```

```python
# Create a array by setting the linear sequence. np.arange(start, stop, increment)
np.arange(0,100,10)
```

```python
# Create a array of 11 values evenly spaced between 0 to 100. np.linspace(start, stop, frequency)
np.linspace(0,100,11)
```

```python
# Create a array of random values between 0 and 1
np.random.random((3,3))
```

```python
# Create a array of normally distributed random values. np.random.normal(mean, std, matrix)
np.random.normal(3,5,(3,3))
```

```python
# Create a array of randome int. np.random.randint(a, b ,(i,j)) number between a and b in i x j matrix
np.random.randint(0,10,(3,3))
```
